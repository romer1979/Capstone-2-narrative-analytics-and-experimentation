```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
client_satisfaction = pd.read_csv('C:/Users/rabia/Desktop/Thinkful/Client_Satisfaction.csv')
client_satisfaction.head()
```

```python
client_satisfaction = client_satisfaction.dropna(subset=['Score'])
client_satisfaction=client_satisfaction[~client_satisfaction['SurveyTemplateName'].str.contains('Ongoing Project')]
client_satisfaction=client_satisfaction[~client_satisfaction['SurveyTemplateName'].str.contains('Pre-Construction')]
```

```python
stage = client_satisfaction.groupby(['SurveyTemplateName']).count()
stage
```

```python
category = client_satisfaction.groupby(['Category']).count()
category
```

```python
accuracy = client_satisfaction.loc[(client_satisfaction['Category'] == 'Accuracy')]
budget = client_satisfaction.loc[(client_satisfaction['Category'] == 'Budget')]
helpfulness = client_satisfaction.loc[(client_satisfaction['Category'] == 'Helpfulness')]
quality = client_satisfaction.loc[(client_satisfaction['Category'] == 'Quality')]
responsiveness = client_satisfaction.loc[(client_satisfaction['Category'] == 'Responsiveness')]
safety = client_satisfaction.loc[(client_satisfaction['Category'] == 'Safety')]
schedule = client_satisfaction.loc[(client_satisfaction['Category'] == 'Schedule')]
```

```python
sns.set(style="whitegrid")
g = sns.pointplot(x="SurveyTemplateName", y="Score", data = accuracy,
                   height=6, kind="strip", palette="pastel",ci=95,dodge=True, join=False)
g.set(ylabel='Score')
g.set(xlabel="")
plt.title('Pointplot: Accuracy Score at Different Projects Stages')
plt.show()
```
